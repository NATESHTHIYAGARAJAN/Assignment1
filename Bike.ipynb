```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.bikewale.com/best-bikes-in-india/"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
name=soup.select(".margin-bottom10 .bw-ga")
```

```python
name
```

```python
price=soup.select("#divSearchResult .font18")
```

```python
price
```

```python
rating=soup.select(".rate-star-color")
```

```python
rating
```

```python
specs=soup.select("#divSearchResult .margin-bottom15 span")
```

```python
specs
```

```python
import pandas as pd
```

```python
bikes=pd.DataFrame(name,columns=['Bike Name'])
bikes["Price"]=price
bikes["Rating"]=rating
bikes["Specs"]=specs
```

```python
bikes
```

```python
bikes.to_csv('D:/Devilrider/bikes.csv')
```
